```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
t = np.arange(500)
y1 = np.sin(t/8)
y2 = np.cos(t/8) + np.random.random(500)/4
```

```python
for i in range(1, 11):
    y = np.cos(t/8 + np.random.random(500)/i)
    fname = 'data/data_%d.txt' %i
    f = open(fname, 'w')
    for a, b in zip(t, y):
        f.write('%.3f, %.3f\n' %(a, b))
    f.close()
    plt.plot(t, y)
```
